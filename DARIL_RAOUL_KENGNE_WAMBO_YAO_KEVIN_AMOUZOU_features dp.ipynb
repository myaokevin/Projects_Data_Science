```python
# 7) nb_restaurant_meme_annee
# Le nombre de restaurants qui sont ouverts leurs portes dans la même année que le restaurant en question.
# Ici, on considère que la première année d'un restaurant correspond à l'année de la première publication d'un avis sur ce restaurant sur Yelp.
# -----------------------------------------------------------

# Votre code ici:

# the first publication date of avis for each restaurant
first_publication_date = avis.groupby('restaurant_id').agg({'date': 'min'})
first_publication_date = first_publication_date.reset_index()
#transformation of the date to datetime like Y-m-d
first_publication_date["date"] = pd.to_datetime(first_publication_date["date"], format='%Y-%m-%d')
first_publication_date['annee'] = first_publication_date['date'].dt.year
# we have to count the number of restaurants that are opened in the same year as x
def count_restaurants_opened_in_the_same_year(restaurant):
    return len(first_publication_date[first_publication_date['annee'] == restaurant['annee']])

first_publication_date['nb_restaurant_meme_annee'] = first_publication_date.apply(count_restaurants_opened_in_the_same_year, axis=1)
first_publication_date



```

## 9) tendance_etoiles

```python
first_publication_date
```

```python
# 9) tendance_etoiles
# La différence entre la moyenne des étoiles de la dernière année et la moyenne des étoiles de la première année d'un restaurant.
# Ici, on considère la première année d'un restaurant correspond à l'année de la première publication d'un avis sur ce restaurant sur Yelp.
# -----------------------------------------------------------

# Votre code ici:

# -----------------------------------------------------------
#liste d'avis de la premiere annee par restaurant
first_publication_date_with_avis = avis.merge(first_publication_date, on='restaurant_id', how='inner')
first_publication_date_with_avis

```

### Calcul de la moyenne par restaurant pour la premiere annee

```python
#transformation de la datede chaque avis en datetime afin de pouvoir extraire l'annee
first_publication_date_with_avis["date_x"] = pd.to_datetime(first_publication_date_with_avis["date_x"], format='%Y-%m-%d')
#selection des avis publies la meme annee que la premiere annee de publication
first_publication_date_with_avis = first_publication_date_with_avis[first_publication_date_with_avis['date_x'].dt.year == first_publication_date_with_avis['annee']]
first_publication_date_with_avis
```

```python
first_publication_date_with_avis = first_publication_date_with_avis.rename(columns={'annee': 'first_year'})
#calcul de la moyenne des etoiles par restaurant 
first_publication_date_with_avis_mean = first_publication_date_with_avis.groupby('restaurant_id').agg({'etoiles': 'mean'})
first_publication_date_with_avis_mean = first_publication_date_with_avis_mean.reset_index()
first_publication_date_with_avis_mean = first_publication_date_with_avis_mean.rename(columns={'etoiles': 'etoiles_mean_first_year'})
first_publication_date_with_avis_mean
```

### Calcul de la moyenne des avis pour la derniere annee

```python

# last publication date of avis for each restaurant
last_publication_date = avis.groupby('restaurant_id').agg({'date': 'max'})
last_publication_date = last_publication_date.reset_index()
#transformation of the date to datetime like Y-m-d
last_publication_date["date"] = pd.to_datetime(last_publication_date["date"], format='%Y-%m-%d')
last_publication_date['last_year'] = last_publication_date['date'].dt.year
last_publication_date_with_avis = avis.merge(last_publication_date, on='restaurant_id', how='inner')
last_publication_date_with_avis
```

```python
#transformation de la datede chaque avis en datetime afin de pouvoir extraire l'annee
last_publication_date_with_avis["date_x"] = pd.to_datetime(last_publication_date_with_avis["date_x"], format='%Y-%m-%d')
#selection des avis publies la meme annee que la derniere annee de publication
last_publication_date_with_avis = last_publication_date_with_avis[last_publication_date_with_avis['date_x'].dt.year == last_publication_date_with_avis['last_year']]
#calcul de la moyenne des etoiles par restaurant
last_publication_date_with_avis_mean = last_publication_date_with_avis.groupby('restaurant_id').agg({'etoiles': 'mean'})
last_publication_date_with_avis_mean = last_publication_date_with_avis_mean.reset_index()
last_publication_date_with_avis_mean  = last_publication_date_with_avis_mean.rename(columns={'etoiles': 'etoiles_mean_last_year'})
last_publication_date_with_avis_mean


```

### calcul de la difference 

```python
#merge of the two dataframes
avis_starts_mean = first_publication_date_with_avis_mean.merge(last_publication_date_with_avis_mean, on='restaurant_id', how='inner')
#add the difference between the mean of the first year and the mean of the last year
avis_starts_mean['tendance_etoiles'] = avis_starts_mean['etoiles_mean_last_year'] - avis_starts_mean['etoiles_mean_first_year']
avis_starts_mean
```

```python
features = pd.read_csv("donnees/features.csv", skipinitialspace=True)
features = features.merge(avis_starts_mean[['restaurant_id', 'tendance_etoiles']], on='restaurant_id', how='left')
features.to_csv('donnees/features.csv', index=False)
features
```

```python

# -----------------------------------------------------------
# 25) chaine
#  Prend 0 ou 1. La valeur 1 indique que le restaurant fait parti d'une chaîne (p. ex. McDonald).
#  On considère un restaurant comme il fait partie d'une chaîne, s’il existe un autre restaurant dans la base de données qui a le même nom.
# -----------------------------------------------------------

# Votre code ici:


#on crée une colonne avec le nom du restaurant en minuscule
restaurants['nom_min'] = restaurants['nom'].str.lower()
restaurants
```

```python

#grouper les restaurants par nom et compter le nombre de restaurant par nom
restaurants_par_nom = restaurants.groupby('nom_min').size()
restaurants_par_nom
```

```python
#fonction qui prend en paramètre le nom du restaurant et retourne 1 si le restaurant fait partie d'une chaine, 0 sinon
def chaine(nom):
    if restaurants_par_nom_map[nom] > 1:
        return 1
    else:
        return 0

#on applique la fonction chaine sur la colonne nom_min
restaurants['chaine'] = restaurants['nom_min'].apply(chaine)
restaurants
```

## 26) nb_heures_ouverture_semaine

```python
# -----------------------------------------------------------
# 26) nb_heures_ouverture_semaine
# Le nombre total d'heures d'ouverture du restaurant par semaine.
# -----------------------------------------------------------

# Votre code ici:
```

```python
horaires  
```

```python
#fonction qui prend en paramete la plage horaire et retourne le nombre d'heures d'ouverture
def nb_heures_ouverture(plage_horaire):

    #we have to check if the plage_horaire is a string or a float
    #if it's a float, it means that the restaurant is closed on this day
    if isinstance(plage_horaire, float):
        return 0
    plage_horaire = str(plage_horaire)
    horaires = plage_horaire.split('-')
    heure_ouverture = horaires[0].split(':')
    heure_fermeture = horaires[1].split(':')
    heure_ouverture_heure = int(heure_ouverture[0])
    heure_ouverture_minute = int(heure_ouverture[1])
    heure_fermeture_heure = int(heure_fermeture[0])
    heure_fermeture_minute = int(heure_fermeture[1])
    #si l'heure de fermeture est supérieur à l'heure d'ouverture, on retourne la différence
    if heure_fermeture_heure > heure_ouverture_heure:
        return (heure_fermeture_heure - heure_ouverture_heure) + (heure_fermeture_minute - heure_ouverture_minute) / 60
    #si l'heure de fermeture est inférieur à l'heure d'ouverture, on retourne 24 -heure d'ouverture + heure de fermeture
    else:
        return 24 - heure_ouverture_heure + heure_fermeture_heure + (heure_fermeture_minute - heure_ouverture_minute) / 60
```

```python
#on crée une colonne avec le nombre d'heures d'ouverture par jour
horaires['nb_heures_ouverture_lundi'] = horaires['lundi'].apply(nb_heures_ouverture)
horaires['nb_heures_ouverture_mardi'] = horaires['mardi'].apply(nb_heures_ouverture)
horaires['nb_heures_ouverture_mercredi'] = horaires['mercredi'].apply(nb_heures_ouverture)
horaires['nb_heures_ouverture_jeudi'] = horaires['jeudi'].apply(nb_heures_ouverture)
horaires['nb_heures_ouverture_vendredi'] = horaires['vendredi'].apply(nb_heures_ouverture)
horaires['nb_heures_ouverture_samedi'] = horaires['samedi'].apply(nb_heures_ouverture)
horaires['nb_heures_ouverture_dimanche'] = horaires['dimanche'].apply(nb_heures_ouverture)

#on crée une colonne avec le nombre total d'heures d'ouverture par semaine pour chaque restaurant
horaires['nb_heures_ouverture_semaine'] = horaires['nb_heures_ouverture_lundi'] + horaires['nb_heures_ouverture_mardi'] + horaires['nb_heures_ouverture_mercredi'] + horaires['nb_heures_ouverture_jeudi'] + horaires['nb_heures_ouverture_vendredi'] + horaires['nb_heures_ouverture_samedi'] + horaires['nb_heures_ouverture_dimanche']
horaires
```

```python
#on merge avec le dataframe features
features = features.merge(horaires[['restaurant_id', 'nb_heures_ouverture_semaine']], on='restaurant_id', how='left')
features.to_csv('donnees/features.csv', index=False)
features = pd.read_csv("donnees/features.csv", skipinitialspace=True)
features
```
